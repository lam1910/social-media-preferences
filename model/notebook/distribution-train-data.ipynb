```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import random
```

```python
data_path = '../../03_Clustering_Marketing.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
def is_float_convertable(x):
    try:
        a = float(x)
        return True
    except:
        return False

mask_can_convert_to_float = df.loc[:, 'age'].apply(lambda x: is_float_convertable(x))

df.loc[~mask_can_convert_to_float, ['age']] = np.NaN
```

```python
dtype_dict = {
    'gradyear': np.float64, 'gender': str, 'age': np.float64, 'NumberOffriends': np.float64, 
    'basketball': np.float64, 'football': np.float64, 'soccer': np.float64, 'softball': np.float64, 
    'volleyball': np.float64, 'swimming': np.float64, 'cheerleading': np.float64, 'baseball': np.float64, 
    'tennis': np.float64, 'sports': np.float64, 'cute': np.float64, 'sex': np.float64, 'sexy': np.float64, 
    'hot': np.float64, 'kissed': np.float64, 'dance': np.float64, 'band': np.float64, 'marching': np.float64, 
    'music': np.float64, 'rock': np.float64, 'god': np.float64, 'church': np.float64, 'jesus': np.float64, 
    'bible': np.float64, 'hair': np.float64, 'dress': np.float64, 'blonde': np.float64, 'mall': np.float64, 
    'shopping': np.float64, 'clothes': np.float64, 'hollister': np.float64, 'abercrombie': np.float64, 
    'die': np.float64, 'death': np.float64, 'drunk': np.float64, 'drugs': np.float64,
}
# change to float64 to avoid warning
df = df.astype(dtype_dict)
```

```python
cols = [
    'gradyear', 'gender', 'age',
    'NumberOffriends', 'volleyball', 'drunk', 'drugs'
]
```

```python
filtered_df = df.dropna()
```

```python
# Plot the histogram
plt.boxplot(filtered_df['age'])

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Age')
plt.title('Distribtion of Age')
plt.ylim(0, 26)

# Show the plot
plt.show()
```

```python
draw_hist = ['gradyear', 'age', 'NumberOffriends']

fig, axes = plt.subplots(1, 3, figsize=(14, 6))
sns.histplot(df['gradyear'], bins=4, ax=axes[0])
axes[0].set_title("Histogram of gradyear")
sns.histplot(df['age'], bins=50, kde=True, ax=axes[1])
axes[1].set_title("Histogram of age")
sns.histplot(df['NumberOffriends'], bins=20, kde=True, ax=axes[2])
axes[2].set_title("Histogram of NumberOffriends")

# Adjust layout
plt.tight_layout()
# Show the plots
plt.show()
```

```python
draw_hist = ['volleyball', 'drunk', 'drugs']

fig, axes = plt.subplots(1, 3, figsize=(14, 6))
sns.histplot(df['volleyball'], bins=10, fill=True, element="poly", ax=axes[0])
axes[0].set_title("Histogram of volleyball post")
sns.histplot(df['drunk'], bins=10, fill=True, element="poly", ax=axes[1])
axes[1].set_title("Histogram of drunk post")
sns.histplot(df['drugs'], bins=10, fill=True, element="poly", ax=axes[2])
axes[2].set_title("Histogram of drugs post")

# Adjust layout
plt.tight_layout()
# Show the plots
plt.show()
```

```python
draw_hist = ['volleyball', 'drunk', 'drugs']

fig, axes = plt.subplots(1, 3, figsize=(14, 6))
sns.kdeplot(df['volleyball'], bw_adjust=3, fill=True, ax=axes[0])
axes[0].set_title("Histogram of volleyball post")
sns.kdeplot(df['drunk'], bw_adjust=3, fill=True, ax=axes[1])
axes[1].set_title("Histogram of drunk post")
sns.kdeplot(df['drugs'], bw_adjust=3, fill=True, ax=axes[2])
axes[2].set_title("Histogram of drugs post")

# Adjust layout
plt.tight_layout()
# Show the plots
plt.show()
```
